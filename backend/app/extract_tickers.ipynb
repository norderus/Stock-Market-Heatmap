```python
import pandas as pd
import json
```

```python
# Data import
data = pd.read_csv('metadata/moex_emitters.csv', sep=';', encoding='windows-1251')
data.head()
```

```python
# Drop NaNs
df_stocks = data[data['SUPERTYPE'] == 'Акции'][['TRADE_CODE', 'ISIN', 'EMITENT_FULL_NAME', 'ISSUE_AMOUNT']].copy()
df_stocks = df_stocks.dropna(subset=['TRADE_CODE'])
df_stocks
```

```python
df_stocks[df_stocks['ISSUE_AMOUNT'] == '3854341416571,43']
```

```python
# Stock extraction
metadata = {}
for _, row in df_stocks.iterrows():
    ticker = row['TRADE_CODE']
    issue_amount = row['ISSUE_AMOUNT']
    
    if pd.notna(issue_amount):
        # Format correction 
        issue_amount = int(float(str(issue_amount).replace(',', '.')))
    else:
        issue_amount = None

    metadata[ticker] = {
        "isin": row['ISIN'],
        "name": row['EMITENT_FULL_NAME'],
        "sector": "Unknown",
        "issue_amount": issue_amount
    }

# Save to JSON
with open('metadata/companies.json', 'w', encoding='utf-8') as f:
    json.dump(metadata, f, ensure_ascii=False, indent=2)

print(f"Сохранено {len(metadata)} записей")
```
